```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
analisis= pd.read_csv("analisis.csv")
analisis
```

```python
# x = analisis.Lote
# y = analisis.pH
fig, ax = plt.subplots()
ax.bar(analisis.Lote, analisis.pH, width=0.5, edgecolor="black", linewidth=1)
#ax.bar(analisis.Lote, analisis.pH)
ax.set_title("Grafico")
plt.xlabel('Lotes')
plt.xlim([0, 5])
plt.ylabel('pH')
plt.ylim([5, 6])
plt.show()
plt.close()
```
